```python
#global variable declare pandren
acc_holder_name = "aaa"
acc_balance=20000
pin=1234
acc_pin=int(input("Enter your pin : ")) #pin input vanguren
if acc_pin==pin:
  print("Correct pin, Welcome!")
  print("1. DEPOSIT")
  print("2, WITHDRAWAL")
  print("3. BALANCE CHECK")
  print("4. EXIT")
  op=int(input("Select anyone option?(1,2,3,4) : "))
  #def vachu functions create pandren
  def withdraw():
    global acc_balance
    amount=float(input("Enter the amount : "))
    if amount <= acc_balance:
      acc_balance -= amount
      print(amount," Withdrawn")
      op=str(input("do you want to know your balance(YES / NO) : ")).upper()
      if op=="YES":
        print("Your available balance is : ",amount)
      else:
        print("Thank you!!")
    else:
      print("Sorry, Insuffecient balance.")
  def deposit():
    global acc_balance
    amount=float(input("Enter the amount : "))
    print(amount," Deposited")
    amount += acc_balance
    print("Your available balance is : ",amount)
  def bal_check():
    global acc_balance
    print("YOUR AVAILABLE BALANCE IS : ",acc_balance)
  def exit():
    print("Thank you!")
  if op==1:
    deposit()
  elif op==2:
    withdraw()
  elif op==3:
    bal_check()
  elif op==4:
    exit()
  else:
    print("Select correct option.")
else:
  print("Wrong pin")
```

```python
#first oru class create pandren
class BANK ():
  #ithula __init__ use panni first global variable declare pannikiren
  def __init__ (self,ahn,ab,pin):
    self.ahn=ahn
    self.ab=ab
    self.pin=pin
  #now im gonna create functions for deposit withdraw and balance check

  #deposit
  def dep(self):
    amt=float(input("Enter your amount : "))
    self.ab += amt
    print(amt," deposited successfully!!")
    op=str(input("do you want to know your balance(YES / NO) : ")).upper()
    if op=="YES":
      print("Your available balance is : ",self.ab)
    else:
      print("Thank you!!")
  def with_draw(self):
    amt=float(input("Enter your amount : "))
    if amt <= self.ab:
      self.ab -= amt
      op=str(input("do you want to know your balance(YES / NO) : ")).upper()
      if op=="YES":
        print("Your available balance is : ",self.ab)
      else:
        print("Thank you!!")
    else:
      print("Sorry, insuffecient balance")
  def bc(self):
    print("Your available balance is : ",self.ab)

  def exiit(self):
    print("Exit, Thank you!")

acc1=BANK("VICKY",20000,2510)
acc2=BANK("RAVI",40000,2520)
acc3=BANK("SUNDARI",60000,2006)

accs=[acc1,acc2,acc3]

pin=int(input("Enter your pin : "))

cur_acc = None

for ac in accs:
  if ac.pin==pin:
    cur_acc=ac
    break
if cur_acc != None:
   print("welcome!!",cur_acc.ahn)
   print("1. Deposit")
   print("2. Withdrawl")
   print("3. Check balance")
   print("4. Exit")
   choi=int(input("Select one option (1,2,3,4) : "))
   if choi==1:
     cur_acc.dep()
   elif choi==2:
     cur_acc.with_draw()
   elif choi==3:
     cur_acc.bc()
   elif choi==4:
     cur_acc.exiit()
   else:
     print("Choose correct option")
else:
  print("Wrong pin!!")
```
